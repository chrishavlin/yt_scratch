```python
import yt
import numpy as np
N = 8
data = {"density": np.random.random((N, N, N))}

ds = yt.load_uniform_grid(
    data,
    [N, N, N],
    bbox=np.array([[-90, 90], [-180, 180], [0.0, 1.0]]),
    geometry=("geographic", ("latitude", "longitude", "z")),
    parameters={'radial_axis': "z"}
)
```

```python
p = ds.proj(("stream", "density"), "z0")
```

```python
ds.create_field_info()
ds.proj(("stream", "density"), "z")
```

```python
yt.SlicePlot(ds, "z", "density")
```

```python
from yt.geometry.coordinates.geographic_coordinates import GeographicCoordinateHandler
```

```python
ordering = ("latitude", "longitude", "level")
gch = GeographicCoordinateHandler(ds, ordering=ordering, radial_axis="level")

ds.coordinates = gch
ds.coordinates.setup_fields(ds.field_info)
```

```python
yt.SlicePlot(ds, "level", "density")
```
