```python
import napari 
import numpy as np 
from napari.utils import nbscreenshot

im2d = np.random.random((100,100))
im3d = np.random.random((100,100, 100))

v = napari.Viewer()
v.dims.ndisplay = 3
v.add_image(im3d, name='3d_image')
v.add_image(im2d, name='2d_image', colormap='magma', opacity=0.7)
v.camera.angles = (0, 10, 30)
v.camera.zoom = 4
nbscreenshot(v)
```

```python
v.add_image(np.random.random((1, 100,100)), 
            name='2dish_image', 
            colormap='magma', 
            opacity=0.7, 
            translate=(50, 0, 0))
nbscreenshot(v)
```
