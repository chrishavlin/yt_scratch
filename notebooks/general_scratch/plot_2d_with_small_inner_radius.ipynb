```python
import yt
import numpy as np 

shp = (64,64,64)
data = {'density': 10**(np.random.random(shp)+5.)}
bbox = np.array([[0.25, 1.], [0, np.pi], [0, 2*np.pi]])
ds = yt.load_uniform_grid(data, shp, 
                          bbox=bbox, 
                          geometry='spherical', 
                          axis_order = ('r', 'theta', 'phi'))
yt.SlicePlot(ds, 'theta', ('stream', 'density'), window_size=(3,3))
```
