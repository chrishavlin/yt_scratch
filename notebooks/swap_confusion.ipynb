# swap confusion

the `swap_axes` and `flip_horizontal` and `flip_vertical` callbacks always resolve to swap first then flip:

```python
import yt
ds = yt.load_sample("IsolatedGalaxy")
```

```python
slc = yt.SlicePlot(ds, 'x', ('enzo', 'Density'))
```

```python
slc.show()
```

add on a `flip_horizontal` to flip what is the image's x axis (the `ds`'s y axis):

```python
slc.flip_horizontal()
```

now if I add on a `swap_axes`, I might expect to just interchange the image's x and y axes as they appear above, but it will actually flip the image's x axis after swapping:

```python
slc.swap_axes()
```

so now we have the `ds`'s z axis flipped instead of y.
