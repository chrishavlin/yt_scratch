```python
import yt
import matplotlib.pyplot as plt 
import numpy as np

ds = yt.load_sample("IsolatedGalaxy")
```

```python
slc = ds.slice(0, 0.5)
frb_wid = 0.1
frb = slc.to_frb(frb_wid, resolution=(400, 400))
d = frb[("enzo", "Density")]

xax = frb.ds.coordinates.x_axis[frb.axis]
yax = frb.ds.coordinates.y_axis[frb.axis]

xn = frb.ds.coordinates.axis_name[xax]
yn = frb.ds.coordinates.axis_name[yax]


plt.imshow(np.log10(d))
plt.xlabel(xn)
plt.ylabel(yn)
plt.show()
```
