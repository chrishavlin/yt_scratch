```python
import yt
import matplotlib.pyplot as plt
import numpy as np 
ds = yt.load_sample("IsolatedGalaxy")
L = [1, 0, 1]  # vector normal to cutting plane
# north_vector = [-1, 1, 0]
W = [0.02, 0.02, 0.02]
c = [0.5, 0.5, 0.5]
N = 512

image = yt.off_axis_projection(ds, 
                               c, 
                               L, 
                               W, 
                               N, 
                               ("gas", "density"), 
                              method='max')
plt.imshow(np.log10(image))
plt.colorbar()
```

```python
import numpy as np
import yt 

bbox = np.array([[0.0, 1.0], [-1., 0.], [0., 1.]])
arr = np.ones((128, 128, 128))
arr[10:20,10:20,10:20] = np.nan
data = dict(ones_with_nans=arr)
ds = yt.load_uniform_grid(data, arr.shape, length_unit="cm", bbox=bbox, nprocs=12)


p = yt.ProjectionPlot(ds, 
                 'z',
                ("stream", "ones_with_nans"), 
                 check_finite=True,
                 # weight_field=("index", "ones"),
                 method='sum')
# p.set_zlim('density', .4,.6)
p.set_log('ones_with_nans', False)
#
```

```python
arr[10:20,10:20,10:20] = np.nan
data = dict(density=arr)
ds = yt.load_uniform_grid(data, arr.shape, length_unit="cm", bbox=bbox, nprocs=12)

p = yt.ProjectionPlot(ds, 
                 'x',
                 ("gas", "density"),                 
                  weight_field=None)
p.set_zlim('density', .4,.6)
```
