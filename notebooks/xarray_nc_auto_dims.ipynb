```python
import numpy as np

T = np.squeeze(ds.T.loc[dict(time=slice(ds.time.values[0]))].values)
```

```python
T.shape
```

```python
import matplotlib.pyplot as plt

plt.contourf(T[-1,:,:],levels=100)
```

```python
data = {"T":T}
```

```python
ytds = yt.load_uniform_grid(data,domain_dimensions=T.shape)
```

```python
sc = yt.create_scene(ytds, "T")
```

```python
T.min()
```

```python
T.max()
```

```python
source = sc[0]

# Set the bounds of the transfer function
source.tfh.set_bounds((1e-5,1e-2))

# set that the transfer function should be evaluated in log space
source.tfh.set_log(True)
```

```python
sc.show()
```

```python
bbox = np.array([
    (ds[dim].values[:].min(),ds[dim].values[:].max()) for dim in ds.T.dims[1:]
])
bbox[1][0]=-89.99
bbox[1][1]=89.99
bbox[2][0]=0.001#-179.99
bbox[2][1]=359.999#179.99
bbox
```

```python
ytds = yt.load_uniform_grid(data,domain_dimensions=T.shape,bbox=bbox,geometry=('geographic',('altitude','latitude','longitude')))
```

```python
cen = ytds.domain_center
yt.SlicePlot(ytds,'altitude','T',center=[900.,cen[1],cen[2]])
```

```python
p = yt.SlicePlot(ytds,'altitude','T',center=[900.,cen[1],cen[2]])

p.set_mpl_projection(('Mollweide', (), {'central_longitude':-110}))#{'central_longitude':-100, 'central_latitude':40}))

p._setup_plots()

p.plots['T'].axes.coastlines()
p.show()
```
