Some examples of using *yt*'s new cm1 netcdf data loader 

```python
import yt 
import os
```

```python
fn = 'testyt.05500.000000.nc'
ds = yt.load(fn)
```

```python
ds.domain_left_edge.to('km')
```

```python
ds.domain_right_edge.to('km')
```

```python
ds.domain_width.to('m')
```

```python
ds.field_list
```

```python
slc = yt.SlicePlot(ds,'x','prespert')
slc.set_axes_unit('m')
slc.show()
```

```python
slc = yt.SlicePlot(ds,'z','prespert')
slc.set_axes_unit('m')
slc.show()
```

```python
slc.zoom(2)
slc.show()
```

```python
slc = yt.SlicePlot(ds,'x','vortmag')
slc.set_axes_unit('m')
slc.show()
```

```python
slc = yt.SlicePlot(ds,'z','vortmag')
slc.set_axes_unit('m')
slc.show()
```

```python
slc.set_log('vortmag',False)
slc.show()
```

```python
sc = yt.create_scene(ds,'vortmag')
source = sc.sources['source_00']
source.set_log(False)
        
tf = yt.ColorTransferFunction((0,7))
tf.add_gaussian(6,.2,[1.0,0.0,0.0,.8])
tf.add_gaussian(5,.2,[0.0,1.0,0.0,.8])
tf.add_gaussian(4,.2,[0.0,0.0,1.0,.5])
source.set_transfer_function(tf)
sc.show()
```

```python
source.set_log(True)        
tf = yt.ColorTransferFunction((-3,2))
tf.add_gaussian(-1,.2,[0.0,0.0,1.0,.01])
tf.add_gaussian(0.2,.05,[0.0,1.0,0.0,.00000001])
tf.add_gaussian(1.5,.4,[1.0,0.0,0.0,1.])
source.set_transfer_function(tf)
sc.render()
sc.show()
```

```python
tf = yt.ColorTransferFunction((-3,2))
tf.add_layers(10, w=0.1)
source.set_transfer_function(tf)
sc.render()
sc.show()
```
