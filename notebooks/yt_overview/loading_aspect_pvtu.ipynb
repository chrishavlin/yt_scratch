```python
fname = 'aspect/output_shell_simple_2d/solution/solution-00003.pvtu'
ds = yt.load(fname)
slc = yt.SlicePlot(ds,'z',('all', 'T'))
slc.set_cmap(('all', 'T'), 'dusk')
slc.annotate_mesh_lines(alpha=0.1, color='w')
slc.show()
```
