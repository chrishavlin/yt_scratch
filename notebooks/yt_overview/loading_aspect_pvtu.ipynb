## mesh element sampling

for unstructured mesh, element evalulations are done with full FEM sampling. 

e.g., given a hexahedral mesh, 

````python
#  3D, linear, hexahedral mesh element has 8 nodes 
from yt.utilities.lib.element_mappings import Q1Sampler3D

hex_sampler = Q1Sampler3D() 
````

For every element, given vertice positions, values at vertices and the positions to sample at:
```python
sampled_values = hex_sampler.sample_at_real_points(vertices, 
                                                   values_at_verts, 
                                                   positions_to_sample)
```

`sample_at_real_point`: 

1. map from real coords to element unit coords
2. sample at element coords 

```python
    @cython.boundscheck(False)
    @cython.wraparound(False)
    @cython.cdivision(True)
    cdef double sample_at_real_point(self,
                                     double* vertices,
                                     double* field_values,
                                     double* physical_x) noexcept nogil:
        cdef double val
        cdef double mapped_coord[4]

        self.map_real_to_unit(mapped_coord, vertices, physical_x)
        val = self.sample_at_unit_point(mapped_coord, field_values)
        return val
```

coordinate mapping from real to unit coords uses newton iterations with Jacobian of element type.
