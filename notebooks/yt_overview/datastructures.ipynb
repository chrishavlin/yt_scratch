```python
import xarray as xr
import yt_xarray
import yt
from yt_xarray import transformations as tf

ds = yt_xarray.open_dataset("IRIS/wUS-SH-2010_percent.nc")
grid_resolution = (32, 32, 32)
gc = tf.GeocentricCartesian(radial_type='depth', r_o=6371., use_neg_lons=True)
ds_yt = tf.build_interpolated_cartesian_ds(
    ds,
    gc,
    fields = 'dvs' ,   
    grid_resolution = grid_resolution, 
    refine_grid=True,    
    refine_max_iters=2000,
    refine_min_grid_size=4,
    refine_by=4,
    interp_method='interpolate',
)
```

```python
import numpy as np
def _slow_nonan(field, data):
    dvs = data['stream', 'dvs']
    dvs[np.isnan(dvs)] = 0.0
    dvs[dvs>0] = 0.0
    return np.abs(dvs)

ds_yt.add_field(
    name=("stream", "slow_nonan"),
    function=_slow_nonan,
    sampling_type="local",
)
```

```python
result = ds_yt.all_data().quantities.max_location(('stream', 'slow_nonan'))
pos = ds_yt.arr(result[1:])
```

```python
c = ds_yt.domain_center.copy()
c[0] = pos[0]
slc = yt.SlicePlot(ds_yt, 'x', ('stream', 'dvs'), center = c)
slc.set_log(("stream", "dvs"), False)
slc.set_cmap(("stream", "dvs"), "magma_r")
slc.set_zlim(("stream", "dvs"), -8, 8)
slc.annotate_cell_edges(color=(1,0,0), alpha=0.25)
slc.annotate_grids(edgecolors=(1,0,1,.5))
slc.set_colorbar_label(("stream", "dvs"), '$\Delta{V_s}$ $\%$')
slc.show()
```

```python
m = ds.index.meshes[0]
con_coords = m.connectivity_coords
con_inds = m.connectivity_indices
```

```python
con_coords.shape, con_inds.shape
```

```python

xvals, yvals
```

```python
el_index = 0
mean_dists = []
for el_index in range(con_inds.shape[0]):
    xvals = con_coords[:, 0][con_inds[el_index]]
    yvals = con_coords[:, 1][con_inds[el_index]]
    yc = yvals.mean()
    xc = xvals.mean()
    mean_dists.append(np.sqrt((xvals - xc)**2 + (yvals - yc)**2).mean())
```

```python
import matplotlib.pyplot as plt 

_ = plt.hist(mean_dists, bins=10)
```
