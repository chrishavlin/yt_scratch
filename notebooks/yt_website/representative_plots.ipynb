# yt summary plots 


## what can yt read? 

> yt supports structured, variable-resolution meshes, unstructured meshes, and discrete or sampled data such as particles.

## what can yt do?

> analyzing and visualizing volumetric data.

From the little widget:

### data operations
* read 
* derived fields (on demand computation)
* geometric subselection `ds.sphere()`

### visualization
* slices
* volume projections
* 3D volume rendering
* binned statistics: ProfilePlots, PhasePlots

### analysis
* extrema
* array-like operations (min, max, etc)

All on larger-than-memory datasets (mixed results in practice depending on data format). 


```python
import yt 
import yt_aspect
```

```python
import yt
ds = yt.load_sample('DeeplyNestedZoom')
slc = yt.SlicePlot(ds,  # dataset
                   'x',  # normal-axis
                   ('gas', 'density'),  # field to plot
                   width=(10,'kpc'),  # width of plot (in physical units)
                   window_size=(3,3,))
slc.annotate_grids(alpha=1, cmap='summer')
slc.set_cmap(('gas', 'density'), 'dusk')
slc.hide_axes()
slc.hide_colorbar()
slc.show()
```
