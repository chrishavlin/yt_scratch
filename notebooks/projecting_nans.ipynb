```python
import yt
import numpy as np

arr = np.random.random(size=(64,64,64))
arr[32, :, :] = np.nan
data = {"density": (arr, "g/cm**3"),}
bbox = np.array([[-1.5, 1.5], [-1.5, 1.5], [-1.5, 1.5]])
ds = yt.load_uniform_grid(data, arr.shape, length_unit="Mpc", bbox=bbox)

prj = ds.proj(("stream", "density"), 'x', method="integrate") # or method="sum"
vals = prj[("stream", "density")]
np.all(np.isnan(vals))
```

```python
prj = yt.ProjectionPlot(ds, 'x', ("stream", "density"))
prj.show()
```

```python
prj = yt.ProjectionPlot(ds, 'y', ("stream", "density"))
prj.show()
```

```python
prj = yt.ProjectionPlot(ds, 'z', ("stream", "density"))
prj.show()
```
