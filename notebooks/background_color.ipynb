```python
import yt
import numpy as np
```

```python

def _densfactor(field, data):
    nanmask = data["density"] < 1e-25
    tmp = data["density"] * 1. 
    tmp[nanmask] = np.nan
    return tmp
```

```python
yt.add_field(("gas", "densfactor"),  _densfactor, 'cell',units = 'g/cm**3', force_override=True)
```

```python

ds = yt.load("IsolatedGalaxy/galaxy0030/galaxy0030")

slc = yt.SlicePlot(ds, 'z', ('gas','densfactor'), center='c')

slc.zoom(30)

slc.show()
```

```python
slc.set_background_color(('gas','densfactor'))
slc.show()
```
