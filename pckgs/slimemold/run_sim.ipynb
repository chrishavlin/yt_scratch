```python
from slime_mold.main import MoldSimulation, _get_bin_ids
import matplotlib.pyplot as plt
import numpy as np 
```

```python
mold2d = MoldSimulation(deposition_factor=.8, 
                        diffusivity=1.5, 
                        n_particles=500, 
                       resolution=(400, 400), 
                       sense_dist=(0.05, .05), 
                       direction_bias=.8, 
                       direction_flip_likelihood=0.1)
plt.imshow(mold2d.deposits)
```

```python
%%time
for _ in range(2000): 
    mold2d.step()
```

```python
from slime_mold.main import MoldSimulation, _get_bin_ids
import matplotlib.pyplot as plt
import numpy as np 

res = (400,400,400)
mold3d = MoldSimulation(deposition_factor=0.8, diffusivity=2., n_particles=500, 
                       resolution=res, 
                       sense_dist=(0.05, 0.05, 0.05),
                       left_edge = (0., 0., 0.),
                       right_edge = (1., 1., 1.),                   
                       direction_bias=0.9)
```

```python
for _ in range(100):
    mold3d.step()
```

```python
plt.figure()
plt.imshow(np.log10(np.sum(mold3d.traces,axis=0)))
plt.colorbar()
```
