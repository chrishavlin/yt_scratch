```python
from slime_mold.main import MoldSimulation, _get_bin_ids
import matplotlib.pyplot as plt
import numpy as np 
```

```python
mold2d = MoldSimulation(deposition_factor=0.8, diffusivity=1., n_particles=1000, 
                       resolution=(400,400), 
                       sense_dist=(0.01, 0.01))
plt.imshow(mold2d.deposits)
```

```python
for _ in range(100): 
    mold2d.step()

plt.figure()
plt.imshow(mold2d.deposits)
plt.colorbar()

plt.figure()
plt.imshow(mold2d.traces)
plt.colorbar()
```
